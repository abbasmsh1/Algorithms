```python
model = LogisticRegression()
Y = np.where(Y > 0.5, 1, 0)  # Convert to binary labels
Y_pred = model.fit_predict(X, Y, lr=1e-5, epochs=100000)
print("Predictions: ", Y_pred)
print("Accuracy: ", accuracy(Y, Y_pred)*100)
```
